```python
# 1. Data Preparation
def load_data(image_dir, label_dir):
    """Load images and masks with proper pairing"""
    image_paths = sorted([os.path.join(r"E:\Cellula Tasks\Week 3\data-20250216T183649Z-001\data\images", f) for f in os.listdir(image_dir) if f.endswith('.tif')])
    label_paths = sorted([os.path.join(r"E:\Cellula Tasks\Week 3\data-20250216T183649Z-001\data\labels", f) for f in os.listdir(label_dir) if f.endswith('.png')])
    
    # Load images and masks
    images = [tiff.imread(img).transpose(1,2,0) for img in image_paths]  # Convert to (128,128,12)
    masks = [cv2.imread(mask, cv2.IMREAD_GRAYSCALE) for mask in label_paths]
    
    # Convert to numpy arrays and normalize
    images = np.array(images) / 255.0
    masks = (np.array(masks) > 127).astype(np.float32)[..., np.newaxis]  # Binarize and add channel
    
    return images, masks
```

```python
def normalize_image(image):
    """Normalize image to 0-1 range with epsilon to avoid division by zero"""
    min_val = np.min(image)
    max_val = np.max(image)
    return (image - min_val) / (max_val - min_val + 1e-8)

def contrast_stretching(image):
    """Apply contrast stretching using 2nd and 98th percentiles"""
    p2, p98 = np.percentile(image, (2, 98))
    return np.clip((image - p2) / (p98 - p2 + 1e-8), 0, 1)

def load_images(image_dir, mask_dir):
    """Improved image loading with filename validation"""
    image_files = sorted(glob.glob(os.path.join(image_dir, "*.tif")))
    mask_files = sorted(glob.glob(os.path.join(mask_dir, "*.png")))
    
    # Verify filename correspondence (assuming similar names)
    paired_files = []
    for img_path in image_files:
        base_name = os.path.basename(img_path).split('.')[0]
        mask_path = next((mp for mp in mask_files if base_name in mp), None)
        if mask_path:
            paired_files.append((img_path, mask_path))
    
    # Load paired data
    images = [tiff.imread(img) for img, _ in paired_files]
    masks = [cv2.imread(mask, cv2.IMREAD_GRAYSCALE) for _, mask in paired_files]
    
    return np.array(images), np.array(masks)

def visualize_bands(image, title="Spectral Bands"):
    """Visualize first 12 bands of a multi-spectral image"""
    plt.figure(figsize=(15, 10))
    for i in range(12):
        plt.subplot(3, 4, i+1)
        plt.imshow(image[:, :, i], cmap='gray')
        plt.title(f"Band {i+1}")
        plt.axis('off')
    plt.suptitle(title)
    plt.show()

# -------------------- Fixed Processing Pipeline --------------------

# Load data
images, masks = load_images(image_dir, mask_dir)

# Verify data shapes
print(f"Original images shape: {images.shape}")
print(f"Original masks shape: {masks.shape}")

# Process images
processed_images = [contrast_stretching(normalize_image(img)) for img in images]
processed_images = np.array(processed_images)  # Convert to numpy array

# Process masks
y_masks = np.array([cv2.resize(mask, (128, 128), interpolation=cv2.INTER_NEAREST) 
                for mask in masks])
y_masks = np.expand_dims(y_masks, axis=-1)  # Add channel dimension
y_masks = (y_masks > 0).astype(np.float32)  # Ensure binary masks

# Verify processed shapes
print(f"Processed images shape: {processed_images.shape}")
print(f"Processed masks shape: {y_masks.shape}")

# Visualize first 4 samples
for i in range(4):
    visualize_bands(processed_images[i], title=f"Processed Bands - Image {i+1}")
    plt.figure()
    plt.imshow(y_masks[i].squeeze(), cmap='gray')
    plt.title(f"Mask {i+1}")
    plt.axis('off')
    plt.show()
```

```python
# Build U-Net model
def build_unet(input_shape=(128, 128, 12)):
    inputs = layers.Input(shape=input_shape)
    
    # Encoder (Contracting Path)
    c1 = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(inputs)
    c1 = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(c1)
    p1 = layers.MaxPooling2D((2, 2))(c1)
    p1 = layers.Dropout(0.5)(p1)  # Dropout to prevent overfitting

    c2 = layers.Conv2D(128, (3, 3), activation='relu', padding='same')(p1)
    c2 = layers.Conv2D(128, (3, 3), activation='relu', padding='same')(c2)
    p2 = layers.MaxPooling2D((2, 2))(c2)
    p2 = layers.Dropout(0.5)(p2)

    c3 = layers.Conv2D(256, (3, 3), activation='relu', padding='same')(p2)
    c3 = layers.Conv2D(256, (3, 3), activation='relu', padding='same')(c3)
    
    # Decoder (Expanding Path)
    u1 = layers.Conv2DTranspose(128, (3, 3), strides=(2, 2), padding='same')(c3)
    u1 = layers.concatenate([u1, c2])
    #u1 = layers.Dropout(0.5)(u1)
    c4 = layers.Conv2D(128, (3, 3), activation='relu', padding='same')(u1)

    u2 = layers.Conv2DTranspose(64, (3, 3), strides=(2, 2), padding='same')(c4)
    u2 = layers.concatenate([u2, c1])
    #u2 = layers.Dropout(0.5)(u2)
    c5 = layers.Conv2D(64, (3, 3), activation='relu', padding='same')(u2)

    # Output Layer
    outputs = layers.Conv2D(1, (1, 1), activation='sigmoid')(c5)

    model = keras.Model(inputs, outputs)
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    
    return model
```
